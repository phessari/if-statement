# IF statement
The if statement in Python is used to only execute block code when a specific condition is true.

1. if statement

2. if else statement

3. elif statement

4. nested if statement

```python
display.Image("if-Statement.png", width=150)
```

__if statement:__


`if (condition is true):
    if body
`

```python
x = 19
if x <= 10:
    print(x)

print('Done!')
```

__if else statement:__

if (condition is True):
    
    if body

else:
    
    else body
    

```python
x = 12

if x < 10:
    print('x is less than 10')
else:
    print('x is greater than 10')
```

# Chained Condition
## elif statement

```python
display.Image("if-elseif-ladder.jpg", width=400)
```

```python
x = -4
if x > 30:
    print('x is greater than 30')
elif x > 10:
    print('x is greater than 10')
elif x > 0:
    print('x is greater than 0')
else:
    print('x is negative')
```

# Nested if condition

```python
display.Image("nested_if.jpg", width=300)
```

```python
x = 12

if x <= 10:
    print('x is less than 10')
else:
    if x % 2 == 0:
        print('x is even')
    print('x is bigger than 10')
```

## The statement can have more than one condition

```python
x = 4
if x > 50:
    print(f'{x} is greater than 50')
elif x > 25 and x <=50:
    print(f'{x} is between 25 and 50')
elif x > 0 and x <= 25:
    print(f'{x} is between 0 and 50')
else:
    print(f'{x} is negative')
```
